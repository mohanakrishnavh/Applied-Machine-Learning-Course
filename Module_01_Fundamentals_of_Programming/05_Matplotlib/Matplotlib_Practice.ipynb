```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([2, 4, 6, 1])
plt.xlabel("Indices")
plt.ylabel("Numbers")
plt.title("Numbers Plot")
plt.show()
```

```python
# plt.plot([--x--], [--y--])
x = np.array([1, 2, 3, 4], dtype=int)
y = x**2
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
plt.plot(x, y, 'ro')
plt.grid()
plt.show()
```

```python
t = np.arange(0,5, 0.2)
plt.plot(t, t**2, 'b--', label='^2', marker='o')
plt.plot(t, t**2.2, 'r--', label='^2.2', marker='s')
plt.plot(t, t**2.5, 'g--', label='^2.5', marker='^')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(x, y, linewidth=5.0)
```

```python
x = np.array([1, 2, 3, 4])
y1 = x**2
y2 = x*2

lines = plt.plot(x, y1, x, y2)
plt.setp(lines[0], color='r', linewidth=2.0)
plt.setp(lines[1], 'color', 'g', 'linewidth', 2.0)
plt.grid()
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0, 5, 0.1)
t2 = np.arange(0, 5, 0.02)

plt.figure(1)
plt.subplot(211)
plt.grid()
plt.plot(t1, f(t1), 'b-')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.show()
```
